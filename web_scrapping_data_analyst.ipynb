```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get('https://in.linkedin.com/jobs/data-analyst-jobs?position=1&pageNum=0')
print(r)
```

```python
r.content
```

```python
soup=BeautifulSoup(r.content,'html.parser')
```

```python
soup.title
```

```python
soup.title.name
```

```python
soup.title.parent.name
```

```python
jobs=[]
campany=[]
links=[]
locations=[]
hiring_status=[]
Seneority_level=[] 
job_function=[]
employment_type=[]
industries=[]
```

```python
for tag in (soup.find_all('h3',attrs={'class':"base-search-card__title"})):
    jobs.append(tag.text.strip())

            
```

```python
for tag in (soup.find_all('a', attrs = {'class':'hidden-nested-link'})):
    campany.append(tag.text.strip())
    
```

```python
len(campany)
```

```python
for tag in (soup.find_all('a', attrs = {'class' : 'hidden-nested-link','href':True})):
    links.append(tag['href'])
```

```python
for tags in (soup.find_all('span', attrs={'class':'job-search-card__location'})):
    locations.append(tags.text.strip())
```

```python
for tag in (soup.find_all('span', attrs={'class':'result-benefits__text'})):
    hiring_status.append(tag.text.strip())
```

```python
soup.find_all('h3',attrs={'class':'description__job-criteria-subheader'})
```

```python
len(links)
```

```python
data = {
    'job_Role':jobs,
    'campany_name':campany,
    'url':links
}
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv('data_analyst')
```
