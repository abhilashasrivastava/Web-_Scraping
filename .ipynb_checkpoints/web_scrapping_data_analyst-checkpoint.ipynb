```python
r=requests.get(u)
print(r)
```

```python
soup=BeautifulSoup(r.content,'html.parser')
```

```python
jobs=[]
campany=[]
links=[]
locations=[]
hiring_status=[]
Seneority_level=[] 
job_function=[]
employment_type=[]
industries=[]
```

```python
for tag in (soup.find_all('h3',attrs={'class':"base-search-card__title"})):
    jobs.append(tag.text.strip())

            
```

```python
len(jobs)
```

```python
jobs
```

```python
for tag in (soup.find_all('a', attrs = {'h4':'hidden-nested-link'})):
    campany.append(tag.text.strip())
```

```python
len(campany)
```

```python
campany
```

```python
for tag in (soup.find_all('a', attrs = {'class' : 'hidden-nested-link','href':True})):
    links.append(tag['href'])
    
    
```

```python
len(links)
```

```python
links
```

```python
for tag in (soup.find_all('span', attrs={'class':'result-benefits__text'})):
    hiring_status.append(tag.text.strip())
```

```python
hiring_status
```

```python
for tags in (soup.find_all('span', attrs={'class':'job-search-card__location'})):
    locations.append(tags.text.strip())
```

```python
locations
```

```python
for link in links:
    r1=request.get(link[0])
    print(r1.status_code)
```

```python
data = {
    'job_Role':jobs,
    'campany_name':campany,
    'url':links
}
```
